```python
# Import necessary libraries
import pandas as pd
from google.colab import files

# Upload the CSV file
print("Please upload your CSV file:")
uploaded = files.upload()

# Load the uploaded CSV file into a DataFrame
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# Function to find and display only duplicate records
def find_duplicates_only(df, id_col, name_col):
    # Identify rows with duplicate ID-name combinations
    duplicates = df[df.duplicated([id_col, name_col], keep=False)]
    return duplicates

# Find duplicates for supplierID and supplierName
supplier_duplicates = find_duplicates_only(data, 'supplierID', 'supplierName')
print("\nRecords with duplicates in 'supplierName' against 'supplierID':")
display(supplier_duplicates)

# Find duplicates for storeID and storeName
store_duplicates = find_duplicates_only(data, 'storeID', 'storeName')
print("\nRecords with duplicates in 'storeName' against 'storeID':")
display(store_duplicates)

# Remove duplicates while keeping the first occurrence
data_cleaned = data.drop_duplicates(subset=['supplierID', 'supplierName'], keep='first')
data_cleaned = data_cleaned.drop_duplicates(subset=['storeID', 'storeName'], keep='first')

# Save the cleaned data to a new CSV file
cleaned_file_name = "cleaned_products_data.csv"
data_cleaned.to_csv(cleaned_file_name, index=False)

print("\nThe duplicates have been removed, and the cleaned data is saved to a new CSV file.")
# Provide a download link for the cleaned file
files.download(cleaned_file_name)
```
